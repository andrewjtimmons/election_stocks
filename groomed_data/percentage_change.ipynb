```python
import pandas as pd
import numpy as np
```

```python
NASDAQ_2016_11_08 = pd.read_csv("../raw_data/NASDAQ_20161108.csv")
NASDAQ_2016_11_09 = pd.read_csv("../raw_data/NASDAQ_20161109.csv")
NYSE_2016_11_08 = pd.read_csv("../raw_data/NYSE_20161108.csv")
NYSE_2016_11_09 = pd.read_csv("../raw_data/NYSE_20161109.csv")

df_list = [
    NASDAQ_2016_11_08,
    NASDAQ_2016_11_09,
    NYSE_2016_11_08,
    NYSE_2016_11_09
]
```

```python
def rename(df):
    df.rename(columns = {
            'Close' : 'Close-' + df.get_value(0, "Date"),
            'Volume' : 'Volume-' + df.get_value(0, "Date"),
    }, inplace = True)

def drop_cols(df):
    df.drop(['Open', 'High', 'Low', 'Date'], axis=1, inplace=True)
```

```python
for df in df_list:
    rename(df)
    drop_cols(df)
```

```python
NASDAQ = pd.merge(NASDAQ_2016_11_08, NASDAQ_2016_11_09, how='inner', on='Symbol')
NASDAQ['exchange'] = 'NASDAQ'
NYSE = pd.merge(NYSE_2016_11_08, NYSE_2016_11_09, how='inner', on='Symbol')
NYSE['exchange'] = 'NYSE'
stocks = pd.concat([NASDAQ, NYSE])
```

```python
def get_price_change(df):
    return (df['Close-09-Nov-2016'] - df['Close-08-Nov-2016'])/df['Close-08-Nov-2016'] * 100
```

```python
stocks['price_change'] = get_price_change(stocks)
```
